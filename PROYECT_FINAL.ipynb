```python
import numpy as np
import matplotlib.pyplot as plt

# Definir función consumo
def C(x):
    return 0.5*x**3 - 3*x**2 + 4*x + 10

# Derivada
def dC(x):
    return 1.5*x**2 - 6*x + 4

x = np.linspace(0, 5, 100)
y = C(x)
dy = dC(x)

plt.plot(x, y, label='Consumo C(x)')
plt.plot(x, dy, label="Derivada C'(x)")
plt.axhline(0, color='black', lw=0.5)
plt.scatter([0.85, 3.15], [C(0.85), C(3.15)], color='red')
plt.legend()
plt.xlabel('Configuración (x)')
plt.ylabel('Consumo de RAM')
plt.title('Consumo de RAM y su derivada')
plt.grid(True)
plt.show()
```
