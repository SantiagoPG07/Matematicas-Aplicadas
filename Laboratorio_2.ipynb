## Ejercicio 1

**Planteamiento:**  
Se desea construir una cerca rectangular con un área de $3600 \, \text{m}^2$, utilizando una pared existente como uno de los lados. ¿Cuáles son las dimensiones del rectángulo que minimizan la cantidad de cerca a usar?

**Solución:**

Sea \( x \) el largo perpendicular a la pared y \( y \) el largo paralelo a la pared.  
Área:   $A = x \cdot y = 3600 \Rightarrow y = $$\frac{3600}{x}$

Cantidad de cerca:  $P = 2x + y = 2x + \frac{3600}{x}$

Minimizar:  

$f(x) = 2x + \frac{3600}{x}$
$\Rightarrow f'(x) = 2 - \frac{3600}{x^2}$
$\Rightarrow f'(x) = 0 \Rightarrow x^2 = 1800 \Rightarrow x = \sqrt{1800}$
$\Rightarrow x \approx 42.43, \quad y = \frac{3600}{x} \approx 84.85$



$\boxed{\text{Dimensiones óptimas: } x \approx 42.43 \, \text{m}, \, y \approx 84.85 \, \text{m}}$

## Ejercicio 2

**Planteamiento:**  
Una hoja rectangular debe contener 54 cm$^2$ de texto. Tiene márgenes de 6 cm arriba y abajo, y 4 cm a los lados. ¿Cuáles deben ser las dimensiones de la hoja para usar la menor cantidad de papel?

**Solución:**

Texto: $xy = 54 \Rightarrow y = \frac{54}{x}$
Dimensiones hoja: $ (x+8) \times (y+12) $

Minimizar área:

$A(x) = (x + 8)\left(\frac{54}{x} + 12\right) = (x + 8)\left(\frac{54 + 12x}{x}\right)
= \frac{12x^2 + 150x + 432}{x}$
$\Rightarrow A'(x) = 12 - \frac{432}{x^2}$
$\Rightarrow x = \sqrt{36} = 6$
$\Rightarrow y = \frac{54}{6} = 9$
$\Rightarrow \text{Dimensiones hoja: } 6+8 = 14, \quad 9+12 = 21$



$ \boxed{\text{Dimensiones óptimas: } 14 \, \text{cm} \times 21 \, \text{cm}} $

## Ejercicio 3

**Planteamiento:**  
Un corral rectangular está delimitado por una pared y se divide internamente en 3 partes con cercas paralelas a uno de los lados. Se desea encerrar 500 m\$^2\$. ¿Cuáles son las dimensiones mínimas de cerca?

**Solución:**

Sea \( x \) el lado paralelo a la pared, y \( y \) el otro. Área: $ xy = 500 \Rightarrow y = \frac{500}{x} $

Cerca: $ C = 2y + 4x = 2\left(\frac{500}{x}\right) + 4x $

Minimizar:

$f(x) = \frac{1000}{x} + 4x
\Rightarrow f'(x) = -\frac{1000}{x^2} + 4
\Rightarrow x^2 = 250 \Rightarrow x = \sqrt{250}
\Rightarrow x \approx 15.81, \quad y = \frac{500}{15.81} \approx 31.62$



$\boxed{\text{Dimensiones óptimas: } x \approx 15.81 \, \text{m}, \, y \approx 31.62 \, \text{m}}$

## Análisis de Costo, Precio y Utilidad Marginal

Dadas las siguientes funciones:

\[
P(x) = 5.00 - 0.002x
\]
\[
C(x) = 3.00 + 1.10x
\]

Vamos a calcular:

1. **El ingreso marginal**
2. **El costo marginal**
3. **La utilidad marginal**
4. **El nivel de producción que maximiza la utilidad total**

---

### 1. Ingreso Marginal

El **ingreso total** $ I(x) $ se obtiene multiplicando el precio por la cantidad:

$
I(x) = P(x) \cdot x = (5.00 - 0.002x) \cdot x = 5x - 0.002x^2
$

Entonces, el **ingreso marginal** $ I_m(x) $ es:

$
I_m(x) = \frac{dI}{dx} = \frac{d}{dx}(5x - 0.002x^2) = 5 - 0.004x
$

---

### 2. Costo Marginal

Dado que:

\[
C(x) = 3.00 + 1.10x
\]

El **costo marginal** \( C_m(x) \) es:

$
C_m(x) = \frac{dC}{dx} = \frac{d}{dx}(3.00 + 1.10x) = 1.10
$

---

### 3. Utilidad Marginal

La **utilidad total** se define como:

$
U(x) = I(x) - C(x)
$

Sustituimos las funciones:

$
U(x) = (5x - 0.002x^2) - (3.00 + 1.10x) = 5x - 0.002x^2 - 3.00 - 1.10x = 3.90x - 0.002x^2 - 3.00
$

Entonces, la **utilidad marginal** \( U_m(x) \) es:

$
U_m(x) = \frac{dU}{dx} = \frac{d}{dx}(3.90x - 0.002x^2 - 3.00) = 3.90 - 0.004x
$

---

### 4. Nivel de producción que maximiza la utilidad

Para maximizar la utilidad, igualamos la **utilidad marginal** a cero:

$
U_m(x) = 3.90 - 0.004x = 0
$

Despejamos $ x $:

$
0.004x = 3.90 \quad \Rightarrow \quad x = \frac{3.90}{0.004} = 975
$

---

###  Conclusión:

- El ingreso marginal es: $ I_m(x) = 5 - 0.004x $
- El costo marginal es constante: $ C_m(x) = 1.10 $
- La utilidad marginal es: $ U_m(x) = 3.90 - 0.004x $
- La utilidad total es máxima cuando se producen **975 unidades**.

##Ejercicio I:

# Análisis y gráfica de \$( f(x) = \dfrac{x}{x+1} \$)

## Paso 1: Análisis preliminar

**(a) Dominio y rango:**

- *Dominio:* $ \mathbb{R} \setminus \{-1\}  (indefinida en  x = -1 ) $
- *Rango:* $ \mathbb{R} \setminus \{1\}  (nunca alcanza el valor 1)$

**(b) Simetría:**

$
f(-x) = \frac{-x}{-x+1} = \frac{x}{x-1} \neq f(x) \text{ ni } -f(x)
$

La función no es par ni impar.

**(c) Intersecciones con los ejes:**

- Con el eje $ x  ( y = 0 ):  (0, 0) $  
- Con el eje $ y  ( x = 0 ):  (0, 0) $

## Paso 2: Análisis con cálculo

**(a) Primera derivada:**

$
f'(x) = \frac{(1)(x+1) - x(1)}{(x+1)^2} = \frac{1}{(x+1)^2} > 0 \quad \forall x \neq -1
$

- Función siempre creciente en su dominio  
- No tiene puntos críticos

**(b) Segunda derivada:**

$
f''(x) = \frac{-2}{(x+1)^3}
$

- Cóncava hacia arriba $( f'' > 0 ) cuando  x < -1 $  
- Cóncava hacia abajo $( f'' < 0) cuando  x > -1$   
- No hay puntos de inflexión (la derivada segunda nunca es cero)

**(c) Asíntotas:**

- *Vertical:* $x = -1 $ (denominador se hace cero)  
- *Horizontal:* \( y = 1 \) (límite cuando \( x \to \pm \infty \))

## Paso 3: Puntos importantes

- Intersección en el origen: $ (0, 0) $
- Comportamiento cerca de asíntotas:

$
\lim_{x \to -1^-} f(x) = +\infty \\
\lim_{x \to -1^+} f(x) = -\infty
$

## Paso 4: Bosquejo de la gráfica

Gráfica de f(x)

**Características destacadas:**

- Discontinuidad infinita en \( x = -1 \)  
- Creciente en todo su dominio  
- Se aproxima a \( y = 1 \) en ambos extremos  
- Cambio de concavidad en \( x = -1 \) (sin ser punto de inflexión)




```python
from IPython.display import display, Math, Latex
import matplotlib.pyplot as plt
import numpy as np

# Generar datos para la gráfica
x = np.linspace(-5, 5, 1000)
x = x[x != -1]  # Excluir punto de discontinuidad
y = x / (x + 1)

# Crear gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = \frac{x}{x+1}$')
plt.axvline(x=-1, color='red', linestyle='--', label='Asíntota vertical x=-1')
plt.axhline(y=1, color='green', linestyle='--', label='Asíntota horizontal y=1')
plt.scatter(0, 0, color='purple', zorder=5, label='Intersección (0,0)')
plt.title("Gráfica de $f(x) = \\frac{x}{x+1}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.ylim(-5, 5)
plt.grid(True)
plt.legend()
plt.show()
```

## Ejercicio II: Gráfica de función par con condiciones específicas

### Condiciones:
1. $f$ es continua
2. $f$ es par (simétrica respecto al eje y)
3. $f(0) = 0$, $f(1) = 2$
4. $f'(x) > 0$ para $x > 0$ (creciente en $(0, +\infty)$)
5. $f''(x) > 0$ para $x > 0$ (cóncava hacia arriba en $(0, +\infty)$)

### Análisis:
- Por ser función par: $f(-x) = f(x)$
- Puntos conocidos: $(0,0)$, $(1,2)$ y por simetría $(-1,2)$
- Comportamiento:
    - Creciente en $(0, +\infty)$ $\Rightarrow$ decreciente en $(-\infty, 0)$ (por ser par)
    - Cóncava hacia arriba en $(0, +\infty)$ $\Rightarrow$ cóncava hacia arriba en $(-\infty, 0)$ (por ser par)
- La función tiene un mínimo en $x=0$

### Bosquejo gráfico:
- Pasa por los puntos $(0,0)$, $(1,2)$ y $(-1,2)$
- Forma de "U" con el vértice en $(0,0)$
- Creciente a la derecha del origen, decreciente a la izquierda
- Siempre cóncava hacia arriba

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 400)
y = x**2 * np.sign(x) * 2  # Ejemplo de función que cumple las condiciones

plt.figure(figsize=(8,6))
plt.plot(x, y, label=r'$f(x)$')
plt.scatter([-1, 0, 1], [2, 0, 2], color='red')
plt.title("Ejercicio II: Gráfica de función par")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

## Ejercicio III: Gráfica con condiciones de derivadas

### Condiciones:
a. $f$ es continua en todas partes  
b. $f(2) = -3$ y $f(6) = 1$  
c. $f'(2) = 0$ y $f'(x) \neq 0$ para $x \neq 2$  
d. $f'(6) = 3$  
e. $f''(6) = 0$; $f''(x) > 0$ para $2 < x < 6$  
f. $f''(x) < 0$ para $x > 6$

### Análisis:
- Puntos conocidos: $(2,-3)$ y $(6,1)$
- Punto crítico en $x=2$ ($f'(2)=0$)
- Comportamiento de la derivada:
    - $f'(6)=3>0$ $\Rightarrow$ función creciente en $x=6$
    - No hay otros puntos críticos ($f'(x)\neq0$ para $x\neq2$)
    - La función es creciente en $(2,+\infty)$
- Comportamiento de la segunda derivada:
    - $f''(x)>0$ en $(2,6)$ $\Rightarrow$ cóncava hacia arriba
    - $f''(6)=0$ $\Rightarrow$ punto de inflexión en $x=6$
    - $f''(x)<0$ en $(6,+\infty)$ $\Rightarrow$ cóncava hacia abajo

### Bosquejo gráfico:
- Pasa por $(2,-3)$ y $(6,1)$
- Mínimo local en $x=2$
- Creciente desde $x=2$ en adelante
- Cóncava hacia arriba entre $x=2$ y $x=6$
- Punto de inflexión en $x=6$
- Cóncava hacia abajo después de $x=6$
- La función tiene una forma de "S" alargada después de $x=2$

```python
# Ejemplo aproximado que cumple las condiciones
x = np.linspace(0, 10, 400)
y = np.piecewise(x,
                [x <= 2, (x > 2) & (x <= 6), x > 6],
                [lambda x: -3 + 0*(x-2)**2,
                 lambda x: -3 + 0.2*(x-2)**3,
                 lambda x: 1 + 3*(x-6) - 0.1*(x-6)**3])

plt.figure(figsize=(8,6))
plt.plot(x, y, label=r'$f(x)$')
plt.scatter([2, 6], [-3, 1], color='red')
plt.title("Ejercicio III: Gráfica con condiciones específicas")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
